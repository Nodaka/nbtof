```python
#@ignore
deg_num: int = 4
scale: float = 10

x = np.arange(-3,7,0.01)
y = 2*x**deg_num + 4*x**(deg_num-1) + np.random.normal(0, scale, x.shape)
```

```python
#@param
x = x
y = y

#@args
params = None

#@kwargs
kwargs = None

#@default
deg_for_fit = 4

#@ignore
plt.scatter(x,y)
```

```python
def fit_func(get_x, *params):
    return_y = np.zeros_like(get_x)
    for i, param in enumerate(params):
        return_y = return_y + np.array(param*get_x**i)
    return return_y
```

```python
popt, pcov = curve_fit(fit_func, x, y, p0=list(np.ones(deg_for_fit+1)))
```

```python
check_y = fit_func(x, *popt)
```
